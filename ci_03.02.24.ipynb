```python
import pandas as pd
df = pd.read_csv('ci.csv')
print(df.to_string())
```

```python
from sklearn.linear_model import LinearRegression
x = df[['year']] #Here we are using two brackets because the logistic regression works under a 2-D GRAPH OR ARRAY
y = df[['per capita income (US$)']]
```

```python
from sklearn.model_selection import train_test_split #train_test_split => divides the data set into two sets for analization
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.33)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
n = int(input())
for i in range(n):
    year,income = map(int,input().split(','))
    print(model.predict([[year,income]]))
        
```
